```python
import pandas as pd
import numpy as np
import math
from datetime import datetime 
import random
import pathlib
from sklearn.utils import shuffle

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', lambda x: '%.02f' % x)

import random, string
n_records = 1000


df_cfo_mis = pd.DataFrame(columns=['aa_subchannel',
'affilate_ind',
'appl_id_num',
'aff_segment',
'aff_portfolio',
'affiliate_channel',
'affiliate_subchannel',
'affiliate_creteria',
'alternate_subchannel',
'aa_int_subchannel',
'cmf_acq_src_cd',
'costco_channel',
'costco_alt_sub_channel',
'costco_int_sub_channel',
'expedia_int_subchannel',
'appv_dt',
'fico_range',
'fsi_ind',
'internate_subchanel',
'orign_cd',
'pid',
'portfolio1',
'portfolio_dan',
'partner',
'product',
'respond_code',
'sbu_product',
'alt_sub_canel',
'brand',
'campaign',
'campaign_year',
'channel',
'date',
'final_sourc_cd',
'portfolio2',
'subchannel',
'source',
'pos11',
'pos13',
'pos15',
'pos16',
'pos17',
'pos10_12',
'pos11_12',
'pos18_19',
'pos1_3',
'pos1_5',
'pos1_7',
'pos4_5',
'pos8_9',
'cntr'])

np.random.seed(25)

possible_values = ['Costco','Custom Cash','Diamond Preferred','Secured Card','Simplicity Value',
                   'AA PLATINUM','Citi Rewards+','Double Cash','ThankYou Premier','MileUp',
                   'Costco BC','AA Business World','AA EXEC','Citi Rewards+ College']

df_cfo_mis['product'] = np.random.choice(possible_values,n_records,p=[.332667,.184133,.119733,.092933,.0632,
                                                                      .0566,.052333,.035867,
                                                                      .017667,.0136,.012267,.008933,.007067,
                                                                      .003])
def partner_con(x):
    if x == "AA Business World" : return "AA BC"
    elif x == "AA EXEC" or x == "AA PLATINUM" : return "AAdvantage"
    elif x == "MileUp" : return "AAdvantage Bronze"
    elif x == "Costco" : return "Costco"
    elif x == "Costco BC" : return "Costco BC"
    else : return "NULL"

df_cfo_mis['partner'] = df_cfo_mis.apply(lambda x : partner_con(x['product']),axis=1)

def portfolio1_con(x):
    if x == "AA Business World" : return "AA BC"
    elif x ==  "AA EXEC" or x ==  "AA PLATINUM" or x ==  "MileUp" : return "AADVANTAGE"
    elif x ==  "Custom Cash" or x ==  "Double Cash" : return "CITICORP CASH"
    elif x ==  "Costco" : return "Costco"
    elif x ==  "Costco BC" : return "Costco BC"
    elif x ==  "Diamond Preferred" or x ==  "Secured Card" or x== "Simplicity Value": return "PPM NON-REWARDS PRODUCT"
    elif x ==  "Citi Rewards+" or x == "Citi Rewards+ College" or x == "ThankYou Premier" : return "REWARDS"
        
    

df_cfo_mis['portfolio1'] = df_cfo_mis.apply(lambda x : portfolio1_con(x['product']),axis=1)

df_cfo_mis['portfolio2'] =  df_cfo_mis['portfolio1'] 


df_cfo_mis['sbu_product'] =  np.where(df_cfo_mis['portfolio1']=='AA BC','AADVANTAGE',df_cfo_mis['portfolio1'])

possible_values = [22647,22645,22646,22648,22649,22650]
df_cfo_mis['appv_dt'] = np.random.choice(possible_values,n_records,p=[.35,.3,.2,.1,.02,.03])

df_cfo_mis['fsi_ind'] = "N"


possible_scores = [">800","761-780","681-700","781-800","661-680","<599",
                   "721-730","741-750","711-720","751-760","731-740",
                   "701-710","641-660","600-640"]

df_cfo_mis['fico_range'] = np.random.choice(possible_scores,n_records,p=[.255,.0928,.0912,.087067,
                                                                        .067267,.058667,.050333,.049,
                                                                        .0484,.046667,.0458,.045667,
                                                                        .039667,.022465])

def portfolio_dan_con(x,y):
    
    if x == "AA BC": return "AA Business Card"
    elif x == "AADVANTAGE" : return "AADVANTAGE" 
    elif x == "CITICORP CASH" : return "CASH"
    elif x == "Costco" : return "Costco Consumer Card"
    elif x == "Costco BC" : return "Costco BC"
    elif x == "PPM NON-REWARDS PRODUCT" and ( y == "Diamond Preferred" or y == "Secured Card"):return "OTHER NON-REWARDS"
    elif x == "PPM NON-REWARDS PRODUCT" and  y == "Simplicity Value": return "SIMPLICITY"
    elif x == "REWARDS" and ( y == "Citi Rewards+" or y == "Citi Rewards+ College"):return "OTHERS"
    elif x == "REWARDS" and ( y == "ThankYou Premier"):return "THANK YOU"
    

df_cfo_mis['portfolio_dan'] = df_cfo_mis.apply(lambda x : portfolio_dan_con(x['portfolio1'],x['product']),axis=1)

def brand_con(x):
    if x == "Costco" or  x == "Costco BC" : return "VISA"
    else: return "MC"

df_cfo_mis['brand'] = df_cfo_mis.apply(lambda x : brand_con(x['portfolio1']),axis=1)

possible_values = [44564,44562,'1/1/2022','1/3/2022',4456]
df_cfo_mis['date'] = np.random.choice(possible_values,n_records,p=[.4,.2,.2,.1,.1])

df_cfo_mis['alt_sub_canel'] = np.NAN # "NULL"


possible_scores = ["BLANK","PH 12","PH 11","PH 1","PH 10","ITA 11","ITA 9","PH 9"]
df_cfo_mis['campaign'] = np.random.choice(possible_scores,n_records,p=[.666667,.199733,.067067,.057067,.0066,.002533,.000267,.000066])


df_cfo_mis['campaign_year'] = df_cfo_mis['campaign'].map({
                    'BLANK':'BLANK',
                    'PH 12':'2021',
                    'PH 11':'2021',
                    'PH 1':'2022',
                    'PH 10':'2021',
                    'ITA 11':'2021',
                    'ITA 9':'2021',
                    'PH 9':'2021'
                    })

df_cfo_mis['source'] = "Blaze"

possible_scores = ["POS","INTERNET NON-RTD","PAPER TM G360","INTERNET","Telemarketing","BATCH TM G360",
                  "INTERNET RTD","RTD TM SALE","BK POS"]
df_cfo_mis['respond_code'] = np.random.choice(possible_scores,n_records,p=[.398895,.262668,.117624,
                                                                          .102219,.044376,.040403,.029358,.004360,.000097])

possible_scores = [520,56,416,257,521,142,408,202,410,89,93
                  ,522,51]
df_cfo_mis['pid'] = np.random.choice(possible_scores,n_records,p=[.465,.121,.104,.09,.066,.062,
                                                                 .026,.017,.017,.014,.008,.006,.004])

possible_scores = ['CW','EA','UW','XU','U1','CX','S3','UZ','BR','S5','BA',
                  'lA','GA','DA','GM','N4']
df_cfo_mis['orign_cd'] = np.random.choice(possible_scores,n_records,p=[.435,.257,.153,.05,.046
                                                                      ,.014,.014,.006,.006,.006,
                                                                      .004,.002,.002,.002,.002,.001])

def conditions_cmf_acq_src_cd(x,length):
    characters = string.ascii_uppercase + string.digits
    return ''.join(random.choice(characters) for _ in range(length))


df_cfo_mis["cmf_acq_src_cd"] = df_cfo_mis.apply(lambda x : conditions_cmf_acq_src_cd(x,11),axis=1)

def create_rand_cod(x):
        characters = string.ascii_uppercase + string.digits
        string_1 = x[0:5] 
        return string_1+''.join(random.choice(characters) for _ in range(14))
    
df_cfo_mis["final_sourc_cd"] = df_cfo_mis.apply(lambda x : create_rand_cod(x['cmf_acq_src_cd']),axis=1)

df_cfo_mis['pos11'] = df_cfo_mis['final_sourc_cd'].str[10]
df_cfo_mis['pos13'] = df_cfo_mis['final_sourc_cd'].str[12]
df_cfo_mis['pos15'] = df_cfo_mis['final_sourc_cd'].str[14]
df_cfo_mis['pos16'] = df_cfo_mis['final_sourc_cd'].str[15]
df_cfo_mis['pos17'] = df_cfo_mis['final_sourc_cd'].str[16]
df_cfo_mis['pos10_12'] = df_cfo_mis['final_sourc_cd'].str[9:12]
df_cfo_mis['pos11_12'] = df_cfo_mis['final_sourc_cd'].str[10:13]
df_cfo_mis['pos18_19'] = df_cfo_mis['final_sourc_cd'].str[17:]
df_cfo_mis['pos1_3'] = df_cfo_mis['final_sourc_cd'].str[:3]
df_cfo_mis['pos1_5'] = df_cfo_mis['final_sourc_cd'].str[:5]
df_cfo_mis['pos1_7'] = df_cfo_mis['final_sourc_cd'].str[:7]
df_cfo_mis['pos4_5'] = df_cfo_mis['final_sourc_cd'].str[3:5]
df_cfo_mis['pos8_9'] = df_cfo_mis['final_sourc_cd'].str[7:9]

df_cfo_mis['cntr'] = 1 

df_cfo_mis['appl_id_num'] = np.random.randint(211011000000000.0,220104000000000.0,n_records)
df_cfo_mis['appl_id_num'] = df_cfo_mis['appl_id_num'].astype('float')

df_cfo_mis['expedia_int_subchannel'] = np.NAN # "NULL"


possible_scores = ['Internet','AltCh','Phase','NPS']
df_cfo_mis['channel'] = np.random.choice(possible_scores,n_records,p=[.333333,.333333,.330533,.002801])


def subchannel_con(x):
    
    if x == "AltCh" : return np.random.choice(['Warehouse','DBCP','Natl Forms','Partner Collateral',
                                              'Other','Citi Channels','Latam','Private Bank Paper APP'],1,p=[0.8209890000000001,.111334,
                                                                                                            .050060,.010813,
                                                                                                            .002203,.000601,
                                                                                                            .00200,.00200])[0]

    else: return "NULL"

df_cfo_mis['subchannel'] = df_cfo_mis.apply(lambda x : subchannel_con(x["channel"]) ,axis=1)

def internate_subchanel_con(x):
    
    if x == "Internet":
        choices = ['External Affiliates','Citicards.com','Partner Sites','Paid Search','Purchase Path',
                  'App-By-Phone','Partner Email','App-by-Phone','AA Paid Search','Prequal','Citi Affiliates','PS email',
                  '0','Paid Media']
        prob = [.455327,.205077,.183690,.076152,.022187,.015391,.011593,.007396,.006996,.005397,.005197,.004797,
               .0006,.0002]
        
        return np.random.choice(choices,1,p=prob)[0]
    
    else:
        return  "NULL"
        
        
    

df_cfo_mis['internate_subchanel'] = df_cfo_mis.apply(lambda x : internate_subchanel_con(x["channel"]) ,axis=1)
df_cfo_mis['affiliate_channel'] = np.NAN # "NULL"
df_cfo_mis['affilate_ind'] = "N"


def add_segment_con(x):
    
    if x == "AltCh":
        value = ["CITIBLUE","Global Client","International","Domestic","BLUE","Domestic Private Bank",
                "BUSINESS CARDS","Citigold","Citi Priority","Private Bank"," Global Private Bank"]
        prob = [0.616459,.25,.03571,.021739,.017081,.009317,.009317,.006211,.003106,.01553,.01553]
        
        return np.random.choice(value,1,p=prob)[0]
        
    else: return "NULL"

df_cfo_mis['aff_segment'] = df_cfo_mis.apply(lambda x : add_segment_con(x["channel"]) ,axis=1)


def aff_portfolio_con(x):
    
    if x == "CITIBLUE" or x == "BUSINESS CARDS" : return "Retail bank"
    else : return "NULL"
    

df_cfo_mis['aff_portfolio'] = df_cfo_mis.apply(lambda x : aff_portfolio_con(x["aff_segment"]) ,axis=1)

def affsubchannel_con(x):
    
    if x == "DBCP": return "DBCP"
    elif x == "Latam" : return "Latam"
    elif x == "Natl Forms" : return "Natl Forms"
    elif x == "Other" : return "Other"
    elif x == "Private Bank Paper APP": return "Private Bank Paper APP"
    else : return "NULL"


df_cfo_mis['affiliate_subchannel'] = df_cfo_mis.apply(lambda x : affsubchannel_con(x["subchannel"]) ,axis=1)


def affiliate_creteria_con(x,y):
    
    if x == "Citi Priority" : return "Citi Priority"
    elif x == "Citigold" : return "CITIGOLD"
    elif x == "Global Client" : return "Global Client"
    elif y == "DBPC" or y == "Natl Forms" : return "NF-NPS"
    elif y == "DBPC" or y == "Other" : return "Other"
    elif x == "CITIBLUE" and y == "Natl Forms" : return "NPS"
    else : return "NULL"

df_cfo_mis['affiliate_creteria'] = df_cfo_mis.apply(lambda x : affiliate_creteria_con(x["aff_segment"],x["affiliate_subchannel"]) ,axis=1)

def alternate_subchannel_con(x):
    
    if x == "NULL" : return np.random.choice(["NULL","CBNA"],1,p=[99/100,1/100])[0]
    else: return "CBNA"
    
    
df_cfo_mis['alternate_subchannel'] = df_cfo_mis.apply(lambda x : alternate_subchannel_con(x["affiliate_creteria"]) ,axis=1)


def aa_int_subchannel_con(x,y):
    
    if (x == "AA BC" or x == "AADVANTAGE") and y == "Internet":
        choices = ["AA Booking Path","AA mobile app","AA Real-Time Benefits",
                  "AA.com","External Booking Path","Solo Email"]
        p = [.459854,.408758,.10073,.023358,.00438,.00292]
        return np.random.choice(choices,1,p=p)[0]
    else:
        return "NULL"

df_cfo_mis['aa_int_subchannel'] = df_cfo_mis.apply(lambda x : aa_int_subchannel_con(x["portfolio1"],x["channel"]) ,axis=1)


df_cfo_mis['aa_subchannel'] = np.NAN # "NULL"

def costco_channel_con(x,y):
    
    if (x == "Costco" or x == "Costco BC") and (y != "NULL"):
        return y
    else: return "NULL"

df_cfo_mis['costco_channel'] = df_cfo_mis.apply(lambda x : costco_channel_con(x["portfolio1"],x["channel"]) ,axis=1)

def costco_alt_sub_channel_con(x):
    choices = ["Tablet","Marketing Brochure LP","Costco Publication LP","Natl Forms","Corporate Liability"]
    p = [0.986007,.01006,.002974,.000719,.00024]
    if x == "AltCh" : return np.random.choice(choices,1,p=p)[0]
    else: return "NULL"


df_cfo_mis['costco_alt_sub_channel'] = df_cfo_mis.apply(lambda x : costco_alt_sub_channel_con(x["costco_channel"]) ,axis=1)


def costco_int_sub_channel_con(x):
    choices = ['Costco.com','Purchase Path','Welcome Email Placements',
              'Welcome Email Placements','Costco Travel.com']
    p = [.615132,.365132,.013158,.003289,.003289]
    if x == "Internet" : return np.random.choice(choices,1,p=p)[0]
    else: return "NULL"


df_cfo_mis['costco_int_sub_channel'] = df_cfo_mis.apply(lambda x : costco_int_sub_channel_con(x["costco_channel"]) ,axis=1)

df_cfo_mis
```
